```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
data.dtypes
```

```python
data.info()
```

```python
data['Make_encode']=lb.fit_transform(data['Make'])
data['Model_encode']=lb.fit_transform(data['Model'])
data['EFT_encode']=lb.fit_transform(data['Engine Fuel Type'])
data['TT_encode']=lb.fit_transform(data['Transmission Type'])
data['DW_encode']=lb.fit_transform(data['Driven_Wheels'])
data['VS_encode']=lb.fit_transform(data['Vehicle Size'])
data['VST_encode']=lb.fit_transform(data['Vehicle Style'])
```

```python
df=data[["Year","Engine HP","Engine Cylinders","Number of Doors","highway MPG","city mpg","Popularity","MSRP","Make_encode","Model_encode","EFT_encode","TT_encode","DW_encode","VS_encode","VST_encode"]]
df.corr()
```

```python
#Creating ML models
```

```python
x=data[['Year','Engine HP',
       'Engine Cylinders',
       'Number of Doors','highway MPG',
       'city mpg', 'Popularity', 'Make_encode', 'Model_encode',
       'EFT_encode', 'TT_encode', 'DW_encode', 'VS_encode', 'VST_encode']]
y=data['MSRP']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=2529)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
lr=LinearRegression()
rft=RandomForestRegressor()
```

```python
lr.fit(x_train,y_train)
```

```python
rft.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
rft.score(x_train,y_train)
```

```python
rft.score(x_test,y_test)
```

```python
import pickle
```

```python
# Save the model to a file using pickle
with open('model.pkl', 'wb') as model_file:
    pickle.dump(rft, model_file)
```
